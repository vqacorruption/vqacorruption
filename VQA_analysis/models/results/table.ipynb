```python
selected_values = {}

# Select values from each dataframe at their specific index
selected_values['QUR'] = df_QUR.loc[df_QUR.index == 'QUR'].values[0]
selected_values['UR'] = df_UR.loc[df_UR.index == 'UR'].values[0]
selected_values['OPAR'] = df_OPAR.loc[df_OPAR.index == 'Percentage'].values[0]
selected_values['ANSL'] = df_ANSL.loc[df_ANSL.index == 'ANSL'].values[0]
selected_values['CEPAR'] = df_CEPAR.loc[df_CEPAR.index == 'Percentage'].values[0]
selected_values['Hallucination_Rate'] = df_AEMR_ALMR_HR.loc[df_AEMR_ALMR_HR.index == 'Hallucination Rate'].values[0]
selected_values['Match_Entity_Rate'] = df_AEMR_ALMR_HR.loc[df_AEMR_ALMR_HR.index == 'Match Entity Rate'].values[0]
selected_values['Match_Layout_Rate'] = df_AEMR_ALMR_HR.loc[df_AEMR_ALMR_HR.index == 'Match Layout Rate'].values[0]
selected_values['QEWR'] = df_QEWR.loc[df_QEWR.index == 'Percentage'].values[0]

for key, value in selected_values.items():
    print(key, value)

# Create df_FINAL from the selected values
df_FINAL = pd.DataFrame(selected_values)
df_FINAL = df_FINAL.transpose()

df_FINAL.index = ['QUR', 'UR', 'OPAR', 'ANSL', 'CEPAR', 'Hallucination_Rate', 'Match_Entity_Rate', 'Match_Layout_Rate', 'QEWR']
df_FINAL.columns = columns
df_FINAL
```

```python
# round the values to 3 decimal places
df_FINAL = df_FINAL.round(3)
```

```python
if "LM" in folder:
    df_FINAL.to_csv(f'{dataset}_FINAL_LM.csv')
else:
    df_FINAL.to_csv(f'{dataset}_FINAL_LLM.csv')
```
